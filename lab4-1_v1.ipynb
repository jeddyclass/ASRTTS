```python
import wget

url = "https://huggingface.co/unsloth/gemma-3-1b-it-GGUF/resolve/main/gemma-3-1b-it-Q4_0.gguf"
output_file = "gemma3_1b_it_q4_0.gguf"
wget.download(url, out=output_file)
```

```python
from llama_cpp import Llama

# Load the Gemma3 1B model
llm = Llama(
    model_path="gemma3_1b_it_q4_0.gguf",  # Replace with full path if needed
    n_ctx=2048,
    n_threads=4  # Adjust depending on your CPU
)

# Prompt format for instruction-tuned models like Gemma
prompt = "[INST] Tell me a bedtime story about a dragon. [/INST]"

# Run inference
output = llm(prompt, max_tokens=100)
print(output["choices"][0]["text"])
```

```python
from llama_cpp import Llama

# Load model (adjust the path if needed)
llm = Llama(model_path="gemma3_1b_it_q4_0.gguf", n_ctx=2048, n_threads=4, chat_format="chatml")

# Structured chat message input
messages = [
    {"role": "user", "content": "Tell me a bedtime story about a dragon."}
]

# Generate response
response = llm.create_chat_completion(messages, max_tokens=100)
print(response["choices"][0]["message"]["content"])
```
